```python
import pandas as pd
```

```python
prod = pd.read_csv("../Clean Dataset/clean_products.csv")
ord = pd.read_csv("../Clean Dataset/clean_orders.csv")
cust = pd.read_csv("../Clean Dataset/clean_customers.csv")
ord_it = pd.read_csv("../Clean Dataset/clean_order_items.csv")
inv = pd.read_csv("../Clean Dataset/clean_inventory.csv")
inv_move = pd.read_csv("../Clean Dataset/clean_inventory_movements.csv")
store = pd.read_csv("../Clean Dataset/clean_stores.csv")
ware = pd.read_csv("../Clean Dataset/clean_warehouses.csv")
```

1. CLEAN PRODUCTS

```python
prod.head()
```

```python
print(prod.dtypes)
```

```python
prod_max_lengths = prod.astype(str).apply(lambda x: x.str.len()).max()
print(prod_max_lengths)
```
