#  Functions

```python
def clean_columns(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    return df
```

```python
def clean_text(df):
    for col in df.select_dtypes(include='object'):
        df[col] = df[col].str.strip()
    return df
```

```python
def convert_to_datetime(df, col):
    df[col] = pd.to_datetime(df[col], errors='coerce')
    return df
```

```python
def preprocess(df):
    df = clean_columns(df)
    df = clean_text(df)
    df = df.drop_duplicates()
    return df
```

# Table-Specific Cleaning

```python
# Customers
cust['signup_date'] = pd.to_datetime(cust['signup_date'], errors='coerce')
cust['city'] = cust['city'].fillna('N/A')
```

```python
# Products
prod = prod[prod['price'] > 0]

prod['price'] = prod['price'].astype(float)

for col in ['brand', 'category']:
    prod[col] = prod[col].str.title()
```

```python
# Orders
ord['order_date'] = pd.to_datetime(ord['order_date'], errors='coerce')

ord['store_id'] = ord['store_id'].str.strip().str.upper()
```

```python
mask = (ord['channel'].isin(['Online', 'App'])) & (ord['store_id'].isna())
ord.loc[mask, 'store_id'] = 'ONLINE'
```

```python
if 'ONLINE' not in store['store_id'].values:
    store = pd.concat([
        store,
        pd.DataFrame([{
            'store_id': 'ONLINE',
            'city': 'N/A',
            'store_type': 'Online'
        }])
    ], ignore_index=True)
```

```python
## Validation where store_id exists
assert ord.loc[ord['store_id'].notna(), 'store_id'].isin(store['store_id']).all()
```

```python
# Order Items
ord_it = ord_it[ord_it['quantity'] > 0]

ord_it['quantity'] = ord_it['quantity'].astype(int)
```

```python
# Inventory
inv['last_updated'] = pd.to_datetime(inv['last_updated'], errors='coerce')
```

```python
# Inventory Movement
inv_move['date'] = pd.to_datetime(inv_move['date'], errors='coerce')
```

```python
inv_move = inv_move[inv_move['change_qty'] != 0]
```

```python
mask2 = inv_move['reason'].isin(['Restock', 'Return'])
inv_move['change_qty'] = np.where(mask2,
                                 inv_move['change_qty'].abs(),
                                 -inv_move['change_qty'].abs())
```

```python
# Stores
store['city'] = store['city'].str.title()


store['store_type'] = store['store_type'].str.title()

store['store_id'] = store['store_id'].str.strip().str.upper()
```

```python
# Warehouses
ware = ware[ware['capacity'] > 0]
```

# Data Engineering

```python
# Creating Revenue Column
ord_it = ord_it.merge(prod[['product_id', 'price']], on='product_id', how='left')
```

```python
ord_it['revenue'] = ord_it['quantity'] * ord_it['price']
```

```python
# Creating Time Feature for order_year
ord['order_year'] = ord['order_date'].dt.year
```

```python
# Creating Time Feature for order_month
ord['order_month'] = ord['order_date'].dt.month
```
