# Data Cleaning

```python
# Removing unnecessary columns (as it contains sensitive information)

df.drop(['name', 'email', 'phone-number', 'credit_card'], axis=1, inplace=True)
```

```python
# Null value check

df.isnull().sum().sort_values(ascending=False)
```

### Note:
Found null values in the columns:
- company
- agent
- country
- children

## Null Value Handling

```python
# column children

df['children'].fillna(0, inplace=True)
```

```python
# column country

df['country'].fillna('Unknown', inplace=True)
```

```python
# column agent

df['agent'].fillna(0, inplace=True)
```

```python
# column company

df['company'].fillna(0, inplace=True)
```

```python
# null value cleaning check

print("Remaining Null Values:\n")

print(df.isnull().sum().sum())
```

## Fixing Datatype

```python
# converting dtype for children, agent & company column.
col = ['children', 'agent', 'company']

for cols in col:
    df[cols] = df[cols].astype(int)
```

```python
# dtype conversion check

df[['children', 'agent', 'company']].dtypes
```

# Adding New Columns

```python
# first of all we need a column for total stay duration

df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
# since we need to know about the total guests.

df['total_guests'] = df['adults'] + df['children'] + df['babies']
```

```python
# total revenue

df['revenue'] = (df['adr'] * df[ 'total_nights'])
```

```python
print("Shape After Cleaning & Feature Engineering:", df.shape)
```

# Outlier Detection

```python
plt.figure(figsize=(10,5))

sns.boxplot(x=df['adr'])

plt.title("ADR Outlier Detection")

plt.show()
```

```python
# removing extreme unrealistic values:
df = df[df['adr'] < 5000]
```

```python
print("Shape After Outlier Removal:", df.shape)
```

```python
df.to_csv("../clean_data/hotel_booking_cleaned.csv", index=False)
```
